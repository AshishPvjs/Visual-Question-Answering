```python
import pickle
import bcolz
vectors = bcolz.open('../6B.300.dat')[:]
words = pickle.load(open('../glove/6B.300_words.pkl', 'rb'))
word2idx = pickle.load(open('../glove/6B.300_idx.pkl', 'rb'))

glove = {w: vectors[word2idx[w]] for w in words}
```

```python
matrix_len = len(vocab_dict)
print(matrix_len)
weights_matrix = np.zeros((matrix_len+1, 300))  #+1 to deal with unk
h=0
words_found=0
words_not_found=[]
for word,i in vocab_dict.items():
    try: 
        weights_matrix[i] = glove[word]
        words_found += 1
    except KeyError:
        h+=1
        words_not_found.append(word)
        weights_matrix[i] = np.random.normal(scale=0.6, size=(300, ))
print(h,words_found)
```
